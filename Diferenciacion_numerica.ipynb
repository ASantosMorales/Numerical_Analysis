## Diferenciación numérica

Utilice:
* a) Aproximación por diferencias finitas hacia adelante.
* b) Aproximación por diferencias finitas hacia atrás.
* c) Aproximación por diferencias finitas centradas

para estimar la primera derivada de

$$f(x) = -0.1x^4 - 0.15x^3 - 0.5x^2 - 0.25x + 1.2$$

en $x=0.5$ utilizando un incremento de $h=0.5$

Se trata de una ecuación ya conocida. Recordando su gráfica:

```python
# Importamos las librerias que se utilizarán
import matplotlib.pyplot as plt
import numpy as np

# Creamos un vector "x" que llenaremos con 100 números
# que van del -1.5 al 1.5.
x = np.linspace(-1.5, 1.5, num = 100)

# Creamos una lista vacía
f_x = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x.append(-0.1*x[i]**4 - 0.15*x[i]**3 - 0.5*x[i]**2 - 0.25*x[i] + 1.2)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

Podemos obtener su derivada de forma analítica y resolver para $x=0.5$

$$f'(x) = -0.4x_i^3 - 0.45x_i^2 - x_i - 0.25$$

y cuando $x=0.5$

$$f'(0.5) = -0.4(0.5)^3 - 0.45(0.5)^2 - (0.5) - 0.25$$

$$f'(0.5) = -0.9125$$

Sabemos que geométricamente una derivada es la pendiente de la recta tangente a la función en un punto. Así que podemos encontrar dicha recta tangente recordando un poco nuestras clases de secundaria.

Para la ecuación en cuestión, si $x=0.5$ entonces

$$y = -0.1(0.5)^4 - 0.15(0.5)^3 - 0.5(0.5)^2 - 0.25(0.5) + 1.2$$

$$y = 0.925$$

Y de esta manera ya tenemos un punto y la pendiente de la recta. Recordamos la ecuación de la recta:

$$y = mx + b$$

Despejamos $b$

$$b = y - mx$$

Sustituyendo

$$b = 0.925 - (-0.9125)(0.5)$$

$$b = 1.38125$$

Es decir

$$y = -0.9125x + 1.38125$$

Graficando

```python
# Creamos una lista vacía
f_x_1_line = []

# Obtenemos los valores de f'(x) para cada "x"
for i in range(len(x)):
    f_x_1_line.append(-0.9125*x[i] + 1.38125)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(0.5, 0.95, color = 'y')
plt.plot(x, f_x_1_line, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

#### Diferencias divididas hacia adelante $x_{i+1}$

$$x_{i} = 0.5\ \ \ \ \ \ \ \ f(x_{i}) = 0.925$$
$$x_{i+1} = 1.0\ \ \ \ \ \ \ f(x_{i+1}) = 0.2$$

Entonces

$$f'(x_i) = \frac{f(x_{i+1}) - f(x_i)}{x_{i+1} - x_i}$$

$$f'(0.5) = \frac{0.2 - 0.925}{1.0 - 0.5} = -1.45$$

Calculando el error relativo porcentual ($|\varepsilon_t|$) contra el valor analítico

$$|\varepsilon_t| = |\frac{-0.9125-(-1.45)}{-0.9125}100\%| = 58.9\%$$

Obteniendo la recta representativa

$$b_{i+1} = y_{i+1} - m_{i+1}x_{i+1} = 0.2 - (-1.45)(1.0) = 1.65$$

Es decir

$$y_{i+1} = -1.45x_{i+1} + 1.65$$

Graficando

```python
# Creamos una lista vacía
f_x_1_forward = []

# Obtenemos los valores de f'(x) para cada "x"
for i in range(len(x)):
    f_x_1_forward.append(-1.45*x[i] + 1.65)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(0.5, 0.95, color = 'y')
plt.plot(x, f_x_1_line, color = 'r')
plt.plot(x, f_x_1_forward, color = 'g')
plt.scatter(1, 0.2, color = 'g')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

#### Diferencias divididas hacia atrás $x_{i-1}$

$$x_{i-1} = 0\ \ \ \ \ \ \ \ f(x_{i-1}) = 1.2$$
$$x_{i} = 0.5\ \ \ \ \ \ \ \ f(x_{i}) = 0.925$$

Entonces

$$f'(x_i) = \frac{f(x_i) - f(x_{i-1})}{x_i - x_{i-1}}$$

$$f'(0.5) = \frac{0.925 - 1.2}{0.5 - 0} = -0.55$$

Calculando el error relativo porcentual ($|\varepsilon_t|$) contra el valor analítico

$$|\varepsilon_t| = |\frac{-0.9125-(-0.55)}{-0.9125}100\%| = 39.7\%$$

Obteniendo la recta representativa

$$b_{i-1} = y_{i-1} - m_{i-1}x_{i-1} = 1.2 - (-0.55)(0) = 1.2$$

Es decir

$$y_{i-1} = -0.55x_{i-1} + 1.2$$

Graficando

```python
# Creamos una lista vacía
f_x_1_backward = []

# Obtenemos los valores de f'(x) para cada "x"
for i in range(len(x)):
    f_x_1_backward.append(-0.55*x[i] + 1.2)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(0.5, 0.95, color = 'y')
plt.plot(x, f_x_1_line, color = 'r')
plt.plot(x, f_x_1_forward, color = 'g')
plt.scatter(1, 0.2, color = 'g')
plt.plot(x, f_x_1_backward, color = 'm')
plt.scatter(0, 1.2, color = 'm')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

#### Diferencias divididas centradas

$$x_{i-1} = 0\ \ \ \ \ \ \ \ f(x_{i-1}) = 1.2$$
$$x_{i+1} = 1.0\ \ \ \ \ \ \ f(x_{i+1}) = 0.2$$

Entonces

$$f'(x_i) = \frac{f(x_{i+1}) - f(x_{i-1})}{x_{i+1} - x_{i-1}}$$

$$f'(0.5) = \frac{0.2 - 1.2}{1 - 0} = -1$$

Calculando el error relativo porcentual ($|\varepsilon_t|$) contra el valor analítico

$$|\varepsilon_t| = |\frac{-0.9125-(-1)}{-0.9125}100\%| = 9.6\%$$

Obteniendo la recta representativa

$$b_{centered} = y_{centered} - m_{centered}x_{centered} = 1.2 - (-1.0)(0) = 1.2$$

Es decir

$$y_{centered} = 1.2 - x_{centered}$$

Graficando

```python
# Creamos una lista vacía
f_x_1_centered = []

# Obtenemos los valores de f'(x) para cada "x"
for i in range(len(x)):
    f_x_1_centered.append(-x[i] + 1.2)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(0.5, 0.95, color = 'y')
plt.plot(x, f_x_1_line, color = 'r')
plt.plot(x, f_x_1_centered, color = 'c')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```
