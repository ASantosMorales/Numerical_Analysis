## Serie de Taylor con función trigonométrica

Utilizar expansiones de Taylor de orden $n$ hasta obtener un error relativo porcentual menor a $0.02\%$ en la aproximación de $f(x) = cos(x)$ en $x_{i+1} = \pi/3$ con base en el valor de $f(x)$ y sus derivadas en $x_i = \pi/4$.

### Solución
Primero graficaremos la función $f(x) = cos(x)$ y localizaremos las solución $f(\pi/3) = cos(\pi/3) = 0.5$.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi/2, num=100)

f_x = []
for i in range(len(x)):
    f_x.append(np.cos(x[i]))

plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.grid()
plt.show()
```

### Aproximación de orden cero $n = 0$

$$f(x_{i+1}) \cong f(x_{i})$$

$$f(\pi/3) \cong f(\pi/4)$$

$$cos(\pi/3) \cong cos(\pi/4)$$

$$0.5 \cong 0.70711$$

Calculando el error relativo porcentual:

$$E_{relativo\ porcentual} = \frac{f(x_{i+1}) - f(x_i)}{f(x_{i+1})}\times100\% = \frac{0.5 - 0.70711}{0.5}\times100\% = -41.42\%$$

Graficando

```python
plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.axhline(1/np.sqrt(2), color = 'r')
plt.scatter(np.pi/3, 1/np.sqrt(2), color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximación de primer orden $n = 1$

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i})$$

$$f(x_{i+1}) \cong cos(x_{i}) - sin(x_i)(x_{i+1}-x_{i})$$

Resolviendo para $x_i = \pi/4$

$$f(x_{i+1}) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(x_{i+1}-\pi/4)$$

Acabamos de obtener la recta para la aproximación de primer orden.

Ahora resolviendo para $x_{i+1} = \pi/3$

$$f(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4)$$

$$cos(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4)$$

$$0.5 \cong 0.52199$$

Calculando el error relativo porcentual:

$$E_{relativo\ porcentual} = \frac{0.5 - 0.52199}{0.5}\times100\% = -4.398\%$$

Graficando

```python
# Creamos una lista vacía
f_x_1 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_1.append(1/np.sqrt(2) - 1/np.sqrt(2)*(x[i] - np.pi/4))

plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.plot(x, f_x_1, color = 'r')
plt.scatter(np.pi/3, 0.52199, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximación de segundo orden $n = 2$

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i}) + \frac{f''(x_{i})}{2!}(x_{i+1}-x_{i})^2$$

$$f(x_{i+1}) \cong cos(x_{i}) - sin(x_i)(x_{i+1}-x_{i}) - \frac{cos(x_i)}{2!}(x_{i+1}-x_{i})^2$$

Resolviendo para $x_i = \pi/4$

$$f(x_{i+1}) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(x_{i+1}-\pi/4) - \frac{1}{2\sqrt2}(x_{i+1}-\pi/4)^2$$

Acabamos de obtener la parábola para la aproximación de segundo orden.

Ahora resolviendo para $x_{i+1} = \pi/3$

$$f(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4) - \frac{1}{2\sqrt2}(\pi/3-\pi/4)^2$$

$$cos(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4) - \frac{1}{2\sqrt2}(\pi/3-\pi/4)^2$$

$$0.5 \cong 0.49775$$

Calculando el error relativo porcentual:

$$E_{relativo\ porcentual} = \frac{0.5 - 0.49775}{0.5}\times100\% = 0.449\%$$

Graficando

```python
# Creamos una lista vacía
f_x_2 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_2.append(1/np.sqrt(2) - 1/np.sqrt(2)*(x[i] - np.pi/4) - 1/(2*np.sqrt(2))*(x[i] - np.pi/4)**2)

plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.plot(x, f_x_2, color = 'r')
plt.scatter(np.pi/3, 0.49775, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximación de tercer orden $n = 3$

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i}) + \frac{f''(x_{i})}{2!}(x_{i+1}-x_{i})^2 + \frac{f'''(x_{i})}{3!}(x_{i+1}-x_{i})^3$$

$$f(x_{i+1}) \cong cos(x_{i}) - sin(x_i)(x_{i+1}-x_{i}) - \frac{cos(x_i)}{2!}(x_{i+1}-x_{i})^2 + \frac{sin(x_i)}{3!}(x_{i+1}-x_{i})^3$$

Resolviendo para $x_i = \pi/4$

$$f(x_{i+1}) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(x_{i+1}-\pi/4) - \frac{1}{2\sqrt2}(x_{i+1}-\pi/4)^2 + \frac{1}{6\sqrt2}(x_{i+1}-\pi/4)^3$$

Acabamos de obtener la función para la aproximación de tercer orden.

Ahora resolviendo para $x_{i+1} = \pi/3$

$$f(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4) - \frac{1}{2\sqrt2}(\pi/3-\pi/4)^2 + \frac{1}{6\sqrt2}(\pi/3-\pi/4)^3$$

$$0.5 \cong 0.49987$$

Calculando el error relativo porcentual:

$$E_{relativo\ porcentual} = \frac{0.5 - 0.49987}{0.5}\times100\% = 0.026\%$$

Graficando

```python
# Creamos una lista vacía
f_x_3 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_3.append(1/np.sqrt(2) - 1/np.sqrt(2)*(x[i] - np.pi/4) - 1/(2*np.sqrt(2))*(x[i] - np.pi/4)**2 + 1/(6*np.sqrt(2))*(x[i] - np.pi/4)**3)

plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.plot(x, f_x_3, color = 'r')
plt.scatter(np.pi/3, 0.49987, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximación de cuarto orden $n = 4$

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i}) + \frac{f''(x_{i})}{2!}(x_{i+1}-x_{i})^2 + \frac{f'''(x_{i})}{3!}(x_{i+1}-x_{i})^3 + \frac{f^(4)(x_{i})}{4!}(x_{i+1}-x_{i})^4$$

$$f(x_{i+1}) \cong cos(x_{i}) - sin(x_i)(x_{i+1}-x_{i}) - \frac{cos(x_i)}{2!}(x_{i+1}-x_{i})^2 + \frac{sin(x_i)}{3!}(x_{i+1}-x_{i})^3 + \frac{cos(x_i)}{4!}(x_{i+1}-x_{i})^4$$

Resolviendo para $x_i = \pi/4$

$$f(x_{i+1}) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(x_{i+1}-\pi/4) - \frac{1}{2\sqrt2}(x_{i+1}-\pi/4)^2 + \frac{1}{6\sqrt2}(x_{i+1}-\pi/4)^3 + \frac{1}{24\sqrt2}(x_{i+1}-\pi/4)^4$$

Acabamos de obtener la función para la aproximación de cuarto orden.

Ahora resolviendo para $x_{i+1} = \pi/3$

$$f(\pi/3) \cong \frac{1}{\sqrt2} - \frac{1}{\sqrt2}(\pi/3-\pi/4) - \frac{1}{2\sqrt2}(\pi/3-\pi/4)^2 + \frac{1}{6\sqrt2}(\pi/3-\pi/4)^3 + \frac{1}{24\sqrt2}(\pi/3-\pi/4)^4$$

$$0.5 \cong 0.5000075$$

Calculando el error relativo porcentual:

$$E_{relativo\ porcentual} = \frac{0.5 - 0.5000075}{0.5}\times100\% = 0.0015\%$$

Graficando

```python
# Creamos una lista vacía
f_x_4 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_4.append(1/np.sqrt(2) - 1/np.sqrt(2)*(x[i] - np.pi/4) - 1/(2*np.sqrt(2))*(x[i] - np.pi/4)**2 + \
                 1/(6*np.sqrt(2))*(x[i] - np.pi/4)**3 + 1/(24*np.sqrt(2))*(x[i] - np.pi/4)**4)

plt.plot(x, f_x)
plt.scatter(np.pi/3, np.cos(np.pi/3), color = 'y')
plt.plot(x, f_x_4, color = 'r')
plt.scatter(np.pi/3, 0.5000075, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```
