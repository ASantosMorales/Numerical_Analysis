### Método Gráfico

Utilice el método gráfico para determinar el coeficiente de arrastre $c$ necesario para que un paracaidista de masa $m = 68.1 kg$ tenga una velocidadde $40 m/s$ después de una caída libre de $t = 10s$. Aceleración de la gravedad $g = 9.8 m/s$.

$$v = \frac{gm}{c}(1-e^{\frac{c}{m}t})$$

#### Solución:
Comenzaremos sustituyendo los datos del problema en la ecuación:

$$40 \frac{m}{s} = \frac{(9.8\frac{m}{s^2})(68.1 kg)}{c}(1-e^{\frac{c}{68.1 kg}(10 s)})$$

Una vez confirmando que las unidades son coherentes y dejando la función expresada en términos de $c$, es decir $f(c)$

$$f(c) = \frac{667.38}{c}(1-e^{-0.146843c})-40$$

Utilizando diversos valores de $c$, debemos encontrar su raíz

```python
import matplotlib.pyplot as plt
import numpy as np
import math

c = np.linspace(1, 20, num = 100)

f_c = []

for i in range(len(c)):
    f_c.append(667.38/c[i]*(1-math.pow(np.e, -0.146843*c[i]))-40)

plt.plot(c, f_c)
plt.grid()
plt.axhline(color = 'k')
plt.axvline(color = 'k')
plt.show()
```

Para dar un valor de $c$ más aproximado al real, nos conviene hacer un zoom en valores que ronden $c = 15 \frac{kg}{s}$

```python
c = np.linspace(14, 16, num = 100)

f_c = []

for i in range(len(c)):
    f_c.append(667.38/c[i]*(1-math.pow(np.e, -0.146843*c[i]))-40)

plt.plot(c, f_c)
plt.grid()
plt.axhline(color = 'k')
plt.show()
```

Para comprobar que el valor de $c$ que escojamos es aproximado podemos sustituirlo en la siguiente ecuación, la cual ya conocemos el resultado con anticipación, $v = 40 \frac{m}{s}$.

$$v_{aprox} = \frac{667.38}{c}(1-e^{-0.146843c})$$

```python
c_aprox = 14.75

print("Si c_approx =", c_aprox, ", la velocidad aproximada (v_aprox) es:", 667.38/c_aprox*(1-math.pow(np.e, -0.146843*c_aprox)))
```
