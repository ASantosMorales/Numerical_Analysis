### Método Gráfico (Ejercicio con funciones trigonométricas)

Localiza las raíces de la siguiente función:

$$f(\alpha) = sin(10\alpha) + cos(3\alpha)$$

Con valores $\alpha_1 = 0$ y $\alpha_2 = 5$

#### Solución

Graficando.

```python
import matplotlib.pyplot as plt
import numpy as np

alpha = np.linspace(0, 5, num = 100)

f_alpha = []

for i in range(len(alpha)):
    f_alpha.append(np.sin(10*alpha[i]) + np.cos(3*alpha[i]))

plt.plot(alpha, f_alpha)
plt.grid()
plt.axhline(color = 'k')
#plt.axvline(color = 'k')
plt.show()
```
