## Serie de Taylor

Recordando que la serie de Taylor se expresa de la siguiente manera:

$$f(x_{i+1}) = f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i}) + \frac{f''(x_{i})}{2!}(x_{i+1}-x_{i})^2 + \frac{f'''(x_{i})}{3!}(x_{i+1}-x_{i})^3 + ... + + \frac{f^{(n)}(x_{i})}{n!}(x_{i+1}-x_{i})^n + R_{n}$$



Resolver utilizando expansiones de Taylor de los 칩rdenes cero hasta tres para aproximar la funci칩n:

$$f(x) = -0.1x^4 - 0.15x^3 - 0.5x^2 - 0.25x + 1.2$$

partiendo de $x_i=0$ estime el valor de $x_{i+1}=1$

Primero se graficar치 con el objetivo de visualizar la funci칩n. Hint: Es posible que no en todos los casos se pueda hacer esto.

```python
# Importamos las librerias que se utilizar치n
import matplotlib.pyplot as plt
import math
import numpy as np

# Creamos un vector "x" que llenaremos con 100 n칰meros
# que van del -2 al 2.
x = np.linspace(-2, 2, num = 100)

# Creamos una lista vac칤a
f_x = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x.append(-0.1*x[i]**4 - 0.15*x[i]**3 - 0.5*x[i]**2 - 0.25*x[i] + 1.2)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximaci칩n de orden cero

Comenzaremos con la aproximaci칩n de orden cero, es decir $n=0$.

Si $n=0$, entonces:

$$f(x_{i+1}) \cong f(x_{i})$$

Recordando que:

$x_i=0$ y $x_{i+1}=1$

Entonces:

$$f(1) \cong f(0)$$

Resolviendo para $f(0)$

$$f(0) = -0.1(0)^4 - 0.15(0)^3 - 0.5(0)^2 - 0.25(0) + 1.2$$

$$f(0) = 1.2$$

Ahora resolviendo para $f(1)$

$$f(1) = -0.1(1)^4 - 0.15(1)^3 - 0.5(1)^2 - 0.25(1) + 1.2$$

$$f(1) = 0.2$$

Calculando el error:

$$E_t = x_{i+1} - x_i = 0.2 - 1.2 = -1$$

Graficando

```python
# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(1, 0.2, color = 'y')
plt.axhline(1.2, color = 'r')
plt.scatter(1, 1.2, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximaci칩n de primer orden $(n = 1)$

Si $n=1$, entonces:

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i})$$

Particularmente para nuestra ecuaci칩n:

$$f(x_{i+1}) \cong -0.1x_i^4 - 0.15x_i^3 - 0.5x_i^2 - 0.25x_i + 1.2 +(-0.4x_i^3 - 0.45x_i^2 - x_i - 0.25)(x_{i+1}-x_{i})$$

Recordando que:

$x_i=0$

$$f(x_{i+1}) \cong -0.1(0)^4 - 0.15(0)^3 - 0.5(0)^2 - 0.25(0) + 1.2 +(-0.4(0)^3 - 0.45(0)^2 - (0) - 0.25)(x_{i+1}- 0)$$

$$f(x_{i+1}) \cong 1.2 - 0.25x_{i+1}$$

Obtuvimos una recta con pendiente que nos ayudar치 a aproximarnos m치s a la soluci칩n.

Ahora recordando que:

$x_{i+1}=1$

$$f(1) = 0.95$$

Calculando el error:

$$E_t = x_{i+1} - x_i = 0.2 - 0.95 = -0.75$$

Graficando

```python
# Creamos una lista vac칤a
f_x_1 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_1.append(1.2 - 0.25*x[i])

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(1, 0.2, color = 'y')
plt.plot(x, f_x_1, color = 'r')
plt.scatter(1, 0.95, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```

### Aproximaci칩n de segundo orden $(n = 2)$

Si $n=2$, entonces:

$$f(x_{i+1}) \cong f(x_{i}) + f'(x_{i})(x_{i+1}-x_{i}) + \frac{f''(x_{i})}{2!}(x_{i+1}-x_{i})^2$$

Para nuestra ecuaci칩n:

$$f(x_{i+1}) \cong -0.1x_i^4 - 0.15x_i^3 - 0.5x_i^2 - 0.25x_i + 1.2 +(-0.4x_i^3 - 0.45x_i^2 - x_i - 0.25)(x_{i+1}-x_i) + \frac{-1.2x_i^2 - 0.9x_i - 1}{2!}(x_{i+1}-x_i)^2$$

Recordando que:

$x_i=0$

$$f(x_{i+1}) \cong -0.1(0)^4 - 0.15(0)^3 - 0.5(0)^2 - 0.25(0) + 1.2 +(-0.4(0)^3 - 0.45(0)^2 - (0) - 0.25)(x_{i+1}- 0) + \frac{-1.2(0)^2 - 0.9(0) - 1}{2}(x_{i+1}-0)^2$$

$$f(x_{i+1}) \cong 1.2 - 0.25x_{i+1} - 0.5x_{i+1}^2$$

Obtuvimos una par치bola que nos ayudar치 a aproximarnos m치s a la soluci칩n.

Ahora recordando que:

$x_{i+1}=1$

$$f(1) = 0.45$$

Calculando el error:

$$E_t = x_{i+1} - x_i = 0.2 - 0.45 = -0.25$$

Graficando

```python
# Creamos una lista vac칤a
f_x_2 = []

# Obtenemos los valores de f(x) para cada "x"
for i in range(len(x)):
    f_x_2.append(1.2 - 0.25*x[i] - 0.5*x[i]**2)

# Graficamos nuestro resultado
plt.plot(x, f_x)
plt.scatter(1, 0.2, color = 'y')
plt.plot(x, f_x_2, color = 'r')
plt.scatter(1, 0.45, color = 'r')
plt.grid()
plt.axvline(color = 'k')
plt.axhline(color = 'k')
plt.show()
```
